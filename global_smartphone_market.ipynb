```python
class Smartphone:
    def _init_(self, brand, model, price_usd, ram_gb, storage_gb,
                 camera_mp, battery_mah, display_size, os, processor):
        self.brand = brand
        self.model = model
        self.price_usd = price_usd
        self.ram_gb = ram_gb
        self.storage_gb = storage_gb
        self.camera_mp = camera_mp
        self.battery_mah = battery_mah
        self.display_size = display_size
        self.os = os
        self.processor = processor

    def _repr_(self):
        return f"{self.brand} {self.model} (${self.price_usd})"
```

```python
import pandas as pd

def load_dataset(path):
    return pd.read_csv(path)

def filter_by_brand(df, brand):
    return df[df["brand"].str.lower() == brand.lower()]

def best_phone_under_budget(df, budget):
    phones = df[df["price_usd"] <= budget]
    return phones.sort_values(by="ram_gb", ascending=False).head(1)

def compare_models(df, model1, model2):
    return df[df["model"].isin([model1, model2])]
```

```python
def market_summary(df):
    return {
        "total_phones": len(df),
        "average_price": df["price_usd"].mean(),
        "brand_distribution": df["brand"].value_counts().to_dict(),
    }

def average_price_by_brand(df):
    return df.groupby("brand")["price_usd"].mean().sort_values()

def os_distribution(df):
    return df["os"].value_counts().to_dict()
```
